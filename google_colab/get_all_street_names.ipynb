```python
import pandas as pd
```

```python
vasserot = pd.read_csv('street_names_vasserot.csv')
```

```python
vasserot.dropna(inplace = True)
```

```python
vasserot['TYPE'] = vasserot['TYPE'].str.upper()
vasserot['NOM'] = vasserot['NOM'].str.upper()
```

```python
pod = pd.read_csv('street_names_pod.csv', sep=';', usecols = ['C_DESI', 'L_VOIE', 'L_LONGMIN'])
```

```python
pod.rename(columns={'C_DESI':'TYPE', 'L_VOIE':'NOM', 'L_LONGMIN':'NOM_ENTIER'},  inplace = True)
```

```python
concat = pd.concat([vasserot, pod])
```

```python
len(concat)
```

```python
concat[concat['NOM_ENTIER'].str.contains('Charonne')]
```

```python
concat.drop_duplicates('NOM_ENTIER', keep="first", inplace = True)
```

```python
len(concat)
```

```python
concat.to_csv('street_names_all.csv')
```
